### preprocesing

```python
import pandas as pd
df = pd.read_excel('top_artis_2022.xlsx')
df
```

```python
df2 = df.drop(['release_date','name','album','artis','popularity'], axis=1)
df2
```

```python
from sklearn.preprocessing import StandardScaler
skaler = StandardScaler()
df2 = pd.DataFrame(skaler.fit_transform(df2), columns=df2.columns)
df2
```

### implementasi model clustering

```python
from sklearn.cluster import KMeans

wws = []
for i in range(1,11):
    model = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=1000, random_state=42)
    model.fit(df2)
    wws.append(model.inertia_)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.plot(range(1,11),wws)
plt.title('grafik elbow')
plt.xlabel('jumlah cluster')
plt.ylabel(wws)
plt.show()
```

```python
kmeans =  KMeans(n_clusters=4, init='k-means++', max_iter=500, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(df2)
```

```python
df2['cluster'] = y_kmeans
df2
```

```python
df2.to_excel(r'hasil_cluster.xlsx', index=False)
```

```python
clus_0 = df2[df2["cluster"]==0]
clus_1 = df2[df2["cluster"]==1]
clus_2 = df2[df2["cluster"]==2]
clus_3 = df2[df2["cluster"]==3]
```

```python
df2.cluster.value_counts()
```

```python
df_result = pd.DataFrame(clus_0.mean(), columns=['Alternative'])
df_result.insert(1, "Pop", clus_1.mean())
df_result.insert(2, "New Age", clus_2.mean())
df_result.insert(3, "Rap", clus_3.mean())
```

```python
df_result
```

```python
plt.scatter(clus_0['energy'], clus_0['lenght'], s=100, c='red', label='Alternative')
plt.scatter(clus_1['energy'], clus_1['lenght'], s=100, c='blue', label='Pop')
plt.scatter(clus_2['energy'], clus_2['lenght'], s=100, c='green', label='New Age')
plt.scatter(clus_3['energy'], clus_3['lenght'], s=100, c='cyan', label='Rap')
plt.title('Rasio cluster berdasarkan energy dan lenght')
plt.xlabel('Energy')
plt.ylabel('Lenght')
plt.legend()
plt.show()
```
